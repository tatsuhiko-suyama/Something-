```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools

# ---------------------------- GLOBALS ----------------------------
K, M = 5, 3  #  K=5, / M=3
mu_tilde = 0.01  # tilde{mu}

# alpha 
alpha_factors = np.linspace(0.0, 2.0, 101)  # : alpha[0,2]  101 

# ---  -------------
r_base = np.array([0.010, 0.015, 0.020, 0.025, 0.030])  # 

Delta_r_m = np.array([  # (KM)  - 
    [-0.005,  0.005,  0.000],  # Asset 0 
    [-0.003,  0.003, -0.002],  # Asset 1 
    [-0.001,  0.001,  0.000],  # Asset 2 
    [ 0.001, -0.001,  0.002],  # Asset 3 
    [ 0.003, -0.003,  0.000]   # Asset 4 
])

sigma_base = np.array([0.15, 0.20, 0.25, 0.30, 0.35])  #  ()

corr_value = 0.5  # 
Corr_base = np.full((K, K), corr_value)
np.fill_diagonal(Corr_base, 1.0)  # 1

V_base = np.diag(sigma_base) @ Corr_base @ np.diag(sigma_base)  # 

# ---------------------------- HELPER FUNCTIONS ----------------------------
def proj_simplex(v):
    """Duchi  v 1."""
    v = np.asarray(v, dtype=float)

    if v.ndim == 0:
        v = np.array([v], dtype=float)
    if v.shape[0] == 1:
        return np.array([1.0], dtype=float)  # If M=1, w must be 1

    tolerance = 1e-9
    if (v >= -tolerance).all() and np.isclose(v.sum(), 1.0, atol=tolerance):
        return v

    u = np.sort(v)[::-1]
    cssv = np.cumsum(u) - 1.0

    rho_candidates_strict = np.where(u - cssv / (np.arange(len(u), dtype=float) + 1.0) > 0)[0]
    rho_idx = rho_candidates_strict[-1] if rho_candidates_strict.size > 0 else 0
    theta = cssv[rho_idx] / (rho_idx + 1.0)
    p = v - theta
    return np.maximum(p, 0.0)

def build_params(alpha):
    """alpha  R_th 2 Sigma_th """
    R_th = r_base[:, None] + alpha * Delta_r_m
    Sigma_th = np.zeros((K, K, M))
    for m in range(M):
        Sigma_th[:, :, m] = V_base + np.outer(R_th[:, m], R_th[:, m])
    return R_th, Sigma_th

# ------------------- QP --------------------------
def inner_opt(w, R, Sigma, tol_feas=1e-7, rcond_pinv=1e-10):
    """ w  pi^T R^m >= mu_tilde """
    Ex = R @ w
    # 
    Vw = sum(w[m] * Sigma[:, :, m] for m in range(M)) - np.outer(Ex, Ex)
    Vw += 1e-10 * np.eye(K)  # 

    try:
        U, S, Vh = np.linalg.svd(Vw)
        if S.size == 0 or (S.size > 0 and np.max(np.abs(S)) < 1e-12):
            Vinv = np.linalg.pinv(Vw, rcond=rcond_pinv)
        else:
            rank = np.sum(S > rcond_pinv * S[0])
            if rank < K:
                Vinv = np.linalg.pinv(Vw, rcond=rcond_pinv)
            else:
                Vinv = np.linalg.inv(Vw)
    except np.linalg.LinAlgError:
        return None, np.inf, Ex, None, None

    best_val = np.inf
    best_pi = None
    best_lam = None
    best_subset = None

    for s in range(1, M + 1):
        for subset_indices_tuple in itertools.combinations(range(M), s):
            subset_indices = list(subset_indices_tuple)
            A = R[:, subset_indices].T
            b = mu_tilde * np.ones(s)

            try:
                Mmat = A @ Vinv @ A.T
                if np.linalg.svd(Mmat)[1].max() < 1e-12:
                    continue

                lam = np.linalg.solve(Mmat, b)
                pi = Vinv @ A.T @ lam
            except np.linalg.LinAlgError:
                continue

            expected_returns_all_models = R.T @ pi
            is_feasible = np.all(expected_returns_all_models >= mu_tilde - tol_feas)
            if is_feasible:
                val = pi @ Vw @ pi
                if np.isfinite(val) and val < best_val - 1e-12:
                    best_val = val
                    best_pi = pi
                    best_lam = lam
                    best_subset = subset_indices

    if best_pi is None:
        return None, np.inf, Ex, None, None

    return best_pi, best_val, Ex, best_lam, best_subset

# ------------------- H  --------------------------
def H_and_grad(w, R, Sigma):
    """
    H(w) 

    Args:
        w: 
        R:  (K x M)
        Sigma: 2 (K x K x M)

    Returns:
        val: H(w) 
        g: w  H 
        pi:  w 
        lam: 
        subset: 
    """
    # 
    pi, obj_val, Ex, lam, subset = inner_opt(w, R, Sigma)

    if pi is None or obj_val == np.inf:
        return np.inf, None, None, None, None

    #  F(pi, w) 
    F_values = np.zeros(M)
    for m in range(M):
        F_values[m] = pi @ Sigma[:, :, m] @ pi - (pi @ R[:, m])**2

    # : F(pi, w)  min_Fobj_val
    g = F_values - obj_val

    #  H(w)  -obj_val
    return -obj_val, g, pi, lam, subset

# -------------------  --------------------------
def optimise_outer(R, Sigma, debug=True):
    """H(w) = min_{pi in A} F(pi, w) over the simplex W."""
    # 
    concentration = np.ones(M) * 0.1  # 
    initial_w_guess = np.random.dirichlet(concentration)
    w = proj_simplex(initial_w_guess)

    lr = 0.1  # 
    tol_grad = 1e-6  # 
    max_iters = 1000  # 

    best_H = -np.inf
    best_w = w.copy()
    best_pi = None
    best_lam = None
    best_subset = None
    converged_iters = max_iters

    current_lr_for_next_iter = lr

    val, g, pi, lam, subset = H_and_grad(w, R, Sigma)

    if val == np.inf or g is None:
        return np.inf, w, None, None, None, 0

    best_H = val
    best_w = w.copy()
    best_pi = pi
    best_lam = lam
    best_subset = subset

    for it in range(1, max_iters + 1):
        w_plus_g = w + g
        w_projected_gradient_check = proj_simplex(w_plus_g)
        projected_grad = w_projected_gradient_check - w
        projected_grad_norm = np.linalg.norm(projected_grad)

        if debug and it % 50 == 0:
            print(f"Iteration {it}: H={val:.6f}, Gradient norm = {projected_grad_norm:.5e}")
            print(f"  w = {w}")
            if pi is not None:
                print(f"  Expected returns: {(R.T @ pi).tolist()}")

        if projected_grad_norm < tol_grad:
            if debug:
                print(f"Converged: grad_norm = {projected_grad_norm:.5e}")
            converged_iters = it
            break

        lr_current_step = current_lr_for_next_iter
        found_accepted_step = False
        while lr_current_step >= 1e-6:
            w_potential = w + lr_current_step * g
            w_new = proj_simplex(w_potential)

            step_size_norm = np.linalg.norm(w_new - w)
            if step_size_norm < 1e-8:
                if debug:
                    print(f"Step size too small: {step_size_norm:.5e}")
                break

            val_new, g_new, pi_new, lam_new, subset_new = H_and_grad(w_new, R, Sigma)

            if val_new == np.inf or g_new is None:
                lr_current_step *= 0.5  # 
                continue

            if val_new >= val - 1e-12:
                w = w_new
                val = val_new
                g = g_new
                pi, lam, subset = pi_new, lam_new, subset_new

                if val > best_H:
                    best_H = val
                    best_w = w.copy()
                    best_pi = pi
                    best_lam = lam
                    best_subset = subset

                current_lr_for_next_iter = min(lr_current_step * 1.2, 1.0)  # 
                found_accepted_step = True
                break
            else:
                lr_current_step *= 0.5  # 

        if not found_accepted_step:
            if debug:
                print("Failed to find acceptable step.")
            converged_iters = it
            break

    if best_H == np.inf:
        return np.inf, best_w, None, None, None, converged_iters

    H_star_final, g_final, pi_star_final, lam_star_final, subset_star_final = H_and_grad(best_w, R, Sigma)

    return H_star_final, best_w, pi_star_final, lam_star_final, subset_star_final, converged_iters

# ------------------- alpha  --------------------------
records = []
infeasible_count = 0
completed_early_count = 0
MAX_OPT_ITERS = 1000

# alpha
num_initializations = 5  # alpha

for i, alpha in enumerate(alpha_factors):
    print(f"Processing alpha = {alpha:.4f} ({i+1}/{len(alpha_factors)})")
    R_th, Sigma_th = build_params(alpha)

    best_H_for_this_alpha = -np.inf
    best_results_for_this_alpha = None

    for init_attempt in range(num_initializations):
        print(f"  Initialization attempt {init_attempt+1}/{num_initializations}")

        H_star, w_star, pi_star, lam_star, subset_star, iters = optimise_outer(R_th, Sigma_th, debug=(init_attempt==0))

        if H_star == np.inf:
            continue

        if H_star > best_H_for_this_alpha:
            best_H_for_this_alpha = H_star
            best_results_for_this_alpha = (H_star, w_star, pi_star, lam_star, subset_star, iters)

    if best_results_for_this_alpha is None:
        infeasible_count += 1
        continue

    H_star, w_star, pi_star, lam_star, subset_star, iters = best_results_for_this_alpha

    if iters < MAX_OPT_ITERS:
        completed_early_count += 1

    active_set = list(subset_star) if subset_star is not None else []
    pi_dot_all_models = (R_th.T @ pi_star) if pi_star is not None else np.full(M, np.nan)

    lam_map_to_models = np.zeros(M)
    if subset_star is not None and lam_star is not None:
        for idx_in_subset, m_idx in enumerate(subset_star):
            lam_map_to_models[m_idx] = lam_star[idx_in_subset]
    lam_out = lam_map_to_models.tolist()

    w_record = w_star.tolist() if w_star is not None else [np.nan] * M
    pi_record = pi_star.tolist() if pi_star is not None else [np.nan] * K
    pi_dot_record = pi_dot_all_models.tolist()

    row = [
        alpha,
        H_star,
        active_set,
        *w_record,
        *pi_record,
        *pi_dot_record,
        *lam_out,
        iters
    ]
    records.append(row)

# 
cols = (
    ["alpha", "H*", "active_set"]
    + [f"w*_{m}" for m in range(M)]       # M columns for w* components
    + [f"pi*_{k}" for k in range(K)]      # K columns for pi* components
    + [f"pi_dot_r{j}" for j in range(M)]  # M columns for pi* . r^j
    + [f"lambda*_{j}" for j in range(M)] # M columns for lambda* corresponding to model j (0 if not active)
    + ["Iterations"]
)

df = pd.DataFrame(records, columns=cols)

# CSV
df.to_csv("alpha_sweep_M3K5_improved.csv", index=False, float_format="%.8e")

# 
plt.figure(figsize=(10, 6))
for m in range(M):
    plt.plot(df['alpha'], df[f'w*_{m}'], label=f'Model {m}')
plt.xlabel('Alpha')
plt.ylabel('Model Weight')
plt.title('Model Weights vs Alpha')
plt.legend()
plt.grid(True)
plt.savefig('weights_vs_alpha.png')

# 
print(df.head())
print(f"Infeasible alphas: {infeasible_count}")
print(f"Early convergence count: {completed_early_count}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools
import time   # optional: for timing

# ---------------------------- GLOBALS ----------------------------
K, M = 5, 4                     #  K M
mu_tilde = 0.03                 #  
alpha_factors = np.linspace(0, 2, 101)  #   [0,2]  101 
# --- Optimizer hyperparameters (globals) ---
MAX_OPT_ITERS = 500        # Upper bound on PG iterations
tol_grad = 1e-8            # Projectedgradient norm tolerance for convergence
step_size_norm_threshold = 1e-10  # Minimum L2 step length accepted in line search

# -------------------  &  --------------------
r_base = np.array([0.010, 0.015, 0.020, 0.025, 0.030])

# r  (PlanB : sectortilted bullish / bearish)
Delta_r_m = 0.001 * np.array([
    [ +3.0, -3.0, +2.5, -2.5],
    [ +2.0, -2.0, +1.5, -1.5],
    [ +1.0, -1.0, +0.5, -0.5],
    [ +1.0, -1.0, +0.5, -0.5],
    [ +1.0, -1.0, +0.5, -0.5],
])  # shape (K, M)

# ----------------  &  ---------------------
sigma_base = np.array([0.15, 0.20, 0.25, 0.30, 0.35])
Delta_sigma_m = 0.06 * np.array([
    [ +1.0, +0.5, +0.8, +0.3],
    [ +0.8, +0.4, +0.6, +0.2],
    [ +0.5, +0.2, +0.4, +0.1],
    [ +0.5, +0.2, +0.4, +0.1],
    [ +0.5, +0.2, +0.4, +0.1],
])  # shape (K, M)

# ---------------- 2 -------------------
def block_corr(rho_in: float, rho_out: float) -> np.ndarray:
    """2sector block correlation matrix
       assets 01  234 """
    C = np.full((K, K), rho_out)
    np.fill_diagonal(C, 1.0)
    sectA, sectB = [0, 1], [2, 3, 4]
    for i in sectA:
        for j in sectA:
            if i != j:
                C[i, j] = rho_in
    for i in sectB:
        for j in sectB:
            if i != j:
                C[i, j] = rho_in
    return C

C_base = block_corr(0.55, 0.55)
# 2
tilde_C_list = [
    block_corr(0.90, 0.10),  # Model 0: extremely high intra / very low cross
    block_corr(0.55, 0.55),  # Model 1: uniform baseline
    block_corr(0.25, 0.75),  # Model 2: crosssector dominant
    block_corr(0.70, 0.20),  # Model 3: strong intra, mild cross
]
beta_corr = 0.8  #   C()= (1- )C_base +   C^m

# ---------- Correlation blending + SPD projection ----------------
def mix_corr(C0: np.ndarray,
             C1: np.ndarray,
             alpha: float,
             beta: float = 0.5,
             eps: float = 1e-4,
             return_lambda_min: bool = False):
    """ SPD """
    C = (1 - beta * alpha) * C0 + beta * alpha * C1
    eigval, eigvec = np.linalg.eigh(C)
    eigval_clip = np.clip(eigval, eps, None)
    C_spd = eigvec @ np.diag(eigval_clip) @ eigvec.T
    D = np.sqrt(np.diag(C_spd))
    C_corr = C_spd / np.outer(D, D)
    if return_lambda_min:
        return C_corr, eigval_clip.min()
    return C_corr
# ---------------------------- HELPER FUNCTIONS ----------------------------
def proj_simplex(v):
    """Duchi  v 1."""
    v = np.asarray(v, dtype=float)

    if v.ndim == 0:
        v = np.array([v], dtype=float)
    if v.shape[0] == 1:
        return np.array([1.0], dtype=float)  # If M=1, w must be 1

    tolerance = 1e-9
    if (v >= -tolerance).all() and np.isclose(v.sum(), 1.0, atol=tolerance):
        return v

    u = np.sort(v)[::-1]
    cssv = np.cumsum(u) - 1.0

    rho_candidates_strict = np.where(u - cssv / (np.arange(len(u), dtype=float) + 1.0) > 0)[0]
    rho_idx = rho_candidates_strict[-1] if rho_candidates_strict.size > 0 else 0
    theta = cssv[rho_idx] / (rho_idx + 1.0)
    p = v - theta
    return np.maximum(p, 0.0)
# -------------------  -------------------------------
def build_params(alpha: float, return_lambda_log: bool = False):
    """
      R_th (KM) 
    2 _th (KKM) 
    return_lambda_log=True  SPD 
    """
    # 
    R_th = r_base[:, None] + alpha * Delta_r_m

    Sigma_th = np.zeros((K, K, M))
    lambda_log = []

    for m in range(M):
        sig_vec = sigma_base + alpha * Delta_sigma_m[:, m]
        diag_sig = np.diag(sig_vec)

        C_alpha, lam_min = mix_corr(
            C_base, tilde_C_list[m], alpha,
            beta=beta_corr, return_lambda_min=True)

        lambda_log.append(lam_min)

        V_m = diag_sig @ C_alpha @ diag_sig
        Sigma_th[:, :, m] = V_m + np.outer(R_th[:, m], R_th[:, m])

    if return_lambda_log:
        return R_th, Sigma_th, lambda_log
    return R_th, Sigma_th
# ------------------- QP --------------------------
#  w  pi^T R^m >= mu_tilde 
# inner_opt 
# M=4, K=5 
def inner_opt(w, R, Sigma, tol_feas=1e-7, rcond_pinv=1e-10):
    """
    Given w, solve min pi^T Vw pi subject to pi^T R^m >= mu_tilde.
    Iterates over all possible active sets.
    R: KxM matrix of expected returns per asset per model
    Sigma: KxKxM tensor of second moment matrices per model
    w: M-dimensional vector of model probabilities
    tol_feas: Tolerance for feasibility check (pi^T R^m >= mu_tilde - tol_feas)
    rcond_pinv: Condition number tolerance for np.linalg.pinv
    Returns optimal pi, min variance, expected return vector, lambda, active subset.
    Returns None, inf, ... if no feasible solution is found for any active set.
    """
    # Calculate the w-weighted average expected return vector E^w[tilde{X}]
    # Ex = sum(w_m * r^m)
    # R is KxM, w is M -> Ex is Kx1
    Ex = R @ w

    # Calculate the w-weighted covariance matrix V^w[tilde{X}]
    # Vw = sum(w_m Sigma^m) - E^w[tilde{X}] E^w[tilde{X}]^t
    Vw = sum(w[m] * Sigma[:, :, m] for m in range(M)) - np.outer(Ex, Ex)

    # Check if Vw is positive definite or at least invertible (numerically)
    # Use SVD for robust inverse and condition check
    try:
        U, S, Vh = np.linalg.svd(Vw)
        # Identify numerically zero singular values based on rcond_pinv
        # Avoid division by zero if S is empty or max(S) is zero
        if S.size == 0 or np.max(S) < 1e-12:
            # Vw is effectively a zero matrix
             # print("Debug: Vw is effectively a zero matrix. Using pinv.")
            Vinv = np.linalg.pinv(Vw, rcond=rcond_pinv)
        else:
            # Calculate inverse using SVD, applying tolerance to singular values
            # s_inv = 1.0 / S; s_inv[S < rcond_pinv * S[0]] = 0.0
            # Vinv = Vh.T @ np.diag(s_inv) @ U.T
            # Using np.linalg.pinv directly is more robust and handles ranks correctly
            Vinv = np.linalg.pinv(Vw, rcond=rcond_pinv)
            # Optional: Check condition number for warning
            # cond_num = S[0] / S[-1] if S[-1] > 0 else np.inf
            # if cond_num > 1.0/rcond_pinv and np.sum(S > rcond_pinv * S[0]) < K:
            #      print(f"Debug: Vw is ill-conditioned (cond={cond_num:.2e}). Using pinv.")


    except np.linalg.LinAlgError:
         # SVD or pinv failed completely
         # print(f"Error: SVD or pseudo-inverse calculation failed for Vw. w={w}")
         return None, np.inf, Ex, None, None # Indicate failure

    best_val = np.inf
    best_pi = None
    best_lam = None
    best_subset = None

    # Iterate over all non-empty subsets of models for active constraints
    for s in range(1, M + 1): # subset size from 1 to M
        for subset_indices_tuple in itertools.combinations(range(M), s):
            subset_indices = list(subset_indices_tuple) # Convert tuple to list

            # Form the active constraint matrix A and vector b
            # A = [r^m1, r^m2, ...]^T for m in subset_indices
            A = R[:, subset_indices].T # (s x K)
            b = mu_tilde * np.ones(s) # (s,)

            # Solve the equality-constrained QP analytically
            # pi = Vinv @ A.T @ (A @ Vinv @ A.T)^{-1} @ b
            # lam = (A @ Vinv @ A.T)^{-1} @ b
            try:
                Mmat = A @ Vinv @ A.T # (s x s)

                # Check if Mmat is singular or ill-conditioned before solving
                U_m, S_m, Vh_m = np.linalg.svd(Mmat)
                # Check if Mmat is effectively zero matrix or singular
                # Use robust rcond for pinv on Mmat
                Mmat_inv = np.linalg.pinv(Mmat, rcond=rcond_pinv)

                # Compute lam = Mmat_inv @ b
                # Using solve instead of pinv for possibly better accuracy if Mmat is well-conditioned
                # But pinv is safer if Mmat is singular, which can happen with certain constraint combinations
                # Let's stick with pinv for robustness as Mmat might be singular even if Vw isn't.
                lam = Mmat_inv @ b # (s,)

                # Compute pi = Vinv A^T lam
                pi = Vinv @ A.T @ lam # (K,)

            except np.linalg.LinAlgError:
                # If any inverse/solve failed unexpectedly
                # print(f"Error: Linear algebra failed for subset {subset_indices}. Skipping.")
                continue # Skip to the next subset
            except ValueError as ve:
                 # Catch potential ValueError from linear algebra ops
                 # print(f"ValueError during solve for subset {subset_indices}: {ve}. Skipping.")
                 continue

            # Check if the found pi is finite and satisfies the non-active constraints
            # pi^T R[:, m] >= mu_tilde for all m NOT in subset_indices
            if not np.all(np.isfinite(pi)):
                 # print(f"Debug: Non-finite pi found for subset {subset_indices}. Skipping.")
                 continue # pi must be finite

            expected_returns_all_models = R.T @ pi # (M,) vector

            # Check feasibility: all elements must be >= mu_tilde (within tolerance)
            is_feasible = np.all(expected_returns_all_models >= mu_tilde - tol_feas)

            # Also check KKT conditions for the active set: lambda must be non-negative
            # This check is required if solving the equality constrained problem.
            # If any lambda for an active constraint is negative, this active set is not the optimal one for the QP.
            kkt_lambda_positive = np.all(lam >= -1e-9) # Tolerance for non-negativity

            if is_feasible and kkt_lambda_positive:
                # This pi is feasible and satisfies KKT for this active set. Calculate the variance.
                val = pi @ Vw @ pi

                # Check if val is finite and this is the best variance found so far.
                # Use a tolerance when comparing floating point values
                if np.isfinite(val) and val < best_val - 1e-12: # Found a strictly better solution
                    best_val    = val
                    best_pi     = pi
                    best_lam    = lam
                    best_subset = subset_indices

    # If after checking all subsets, no feasible pi with non-negative lambda was found
    if best_pi is None:
        # print(f"Debug: No feasible pi found for w={w}. Returning inf.")
        return None, np.inf, Ex, None, None # Indicate infeasibility

    # Final check that the selected best solution is valid
    if not np.all(np.isfinite(best_pi)) or not np.isfinite(best_val):
         # This should ideally not happen if best_pi was set, but defensive check
         # print(f"Warning: Non-finite best_pi or best_val found for w={w} after iterating all subsets. Returning inf.")
         return None, np.inf, Ex, None, None # Indicate infeasibility or error

    return best_pi, best_val, Ex, best_lam, best_subset


# -------------------  H  --------------------------
# H(w) = min_{pi in A} F(pi, w) and its gradient w.r.t. w
def H_and_grad(w, R, Sigma, tol_feas=1e-7, rcond_pinv=1e-10, debug=False):
    """
    Calculates H(w) = min_{pi in A} F(pi, w) and its gradient w.r.t w.
    Based on the Envelope Theorem.
    """
    # Call inner_opt to get the optimal pi*(w) and the value H(w)
    # inner_opt returns: pi, val, Ex, lam, subset
    # Pass down tolerances
    pi, val, Ex, lam, subset = inner_opt(w, R, Sigma, tol_feas=tol_feas, rcond_pinv=rcond_pinv)

    # If inner_opt failed to find a feasible solution, H(w) is effectively infinity
    if val == np.inf:
        if debug: print(f"  H_and_grad: Inner optimization returned infeasible solution.")
        return np.inf, None, pi, lam, subset # Returning None gradient indicates failure

    # Calculate the gradient based on the Envelope Theorem
    # \nabla H(\bm{w})_m = \left. \frac{\partial F(\bm{\pi}, \bm{w})}{\partial w_m} \right|_{\bm{\pi}=\bm{\pi}^*(\bm{w})}
    # F(pi, w) = pi^T Vw pi
    # Vw = sum(w_k Sigma^k) - Ex Ex^T, where Ex = sum(w_k r^k)
    # dF/dw_m = pi^T Sigma^m pi - 2 * (pi^T Ex) * (pi^T r^m)

    # Check if pi is finite before calculating gradient (should be caught by inner_opt returning inf)
    if not np.all(np.isfinite(pi)):
         if debug: print(f"  H_and_grad: Non-finite pi passed to gradient calculation.")
         return val, None, pi, lam, subset # Indicate failure

    piTEx = pi @ Ex # pi^T * (sum w_k r^k)
    grad = np.array([
        pi @ Sigma[:, :, m] @ pi - 2 * (pi @ R[:, m]) * piTEx
        for m in range(M) # Iterate over models 0 to M-1
    ])

    # Check if gradient is valid (not NaN or Inf)
    if not np.all(np.isfinite(grad)):
         if debug: print(f"  H_and_grad: Non-finite gradient calculated: {grad}. Returning None.")
         return val, None, pi, lam, subset # Indicate failure for optimizer

    if debug:
        # print(f"  H_and_grad: w = {w}")
        # print(f"  H_and_grad: pi = {pi}")
        # print(f"  H_and_grad: H(w) = {val:.6e}")
        print(f"  H_and_grad: Calculated gradient g = {grad}. L2 norm = {np.linalg.norm(grad):.2e}")
        # Check constraints feasibility at the found pi
        # exp_rets = R.T @ pi
        # print(f"  H_and_grad: pi.T @ r_m = {exp_rets} (mu_tilde = {mu_tilde})")
        # Check KKT lambda for active set (already done in inner_opt but good for debug)
        # if subset is not None:
        #      print(f"  H_and_grad: Active set indices = {subset}")
        #      print(f"  H_and_grad: Lambda values for active set = {lam}")


    return val, grad, pi, lam, subset


# -------------------  --------------------------
# H(w)  w >= 0, sum(w)=1 
def optimise_outer(R, Sigma, debug=True):
    """
    Maximize H(w) = min_{pi in A} F(pi, w) over the simplex W.
    Uses Projected Gradient Ascent with simple backtracking line search.
    R: KxM matrix
    Sigma: KxKxM tensor
    debug: print debug info (more detailed for this version)
    Returns H_star, w_star, pi_star, lam_star, subset_star, iters.
    Returns inf, w, None, None, None, iters if optimization leads to infeasible state.
    """
    # Initialize with the center of the simplex
    initial_w_guess = np.ones(M) / M
    w = initial_w_guess.copy() # Ensure it's a copy
    if debug: print(f"\nOuter Opt: --- Starting optimization ---")
    if debug: print(f"Outer Opt: Initial w = {w} (center of simplex)")


    # Optimization parameters (using global values)
    lr_init = 0.15 # Initial learning rate for the first iteration's line search starting point
    # tol_grad is global
    # MAX_OPT_ITERS is global

    # Variable to track the best valid solution found
    best_H = -np.inf # Initialize with negative infinity for maximization
    best_w = w.copy() # Initialize best_w with the starting point
    best_pi = None
    best_lam = None
    best_subset = None
    # converged_iters will store the iteration count *when* convergence/stop was detected
    converged_iters = MAX_OPT_ITERS # Default to max_iters if loop completes

    # Simple backtracking line search parameters
    lr_decay = 0.5 # Factor to decrease learning rate in line search
    lr_grow = 1.2 # Factor to increase learning rate for the next iteration's starting lr
    min_lr_line_search = 1e-6 # Minimum learning rate allowed *within* a single line search step
    max_lr_next_iter = 0.5 # Maximum learning rate allowed for the *next* iteration's starting lr

    # Tolerance for value improvement check in line search (Armijo-like condition simplified)
    # We require the new value to be at least the old value minus a small tolerance for numerical stability
    tol_val_improve = 1e-12 # Minimum required non-decrease in H(w) to accept step

    # step_size_norm_threshold is global

    # Use a tighter loop limit for line search to prevent infinite loops if parameters are bad
    max_line_search_steps = 50 # Max steps within one line search

    # Tolerance for numerical stability in inner_opt (passed down)
    tol_inner_feas = 1e-7
    rcond_inner_pinv = 1e-10

    # Get initial value and gradient at the starting point w
    # Pass down tolerances and debug flag for inner evaluation
    # Pass debug flag down to H_and_grad for detailed inner step info if needed
    val, g, pi, lam, subset = H_and_grad(w, R, Sigma, tol_inner_feas, rcond_inner_pinv, debug=debug)

    # If initial state is infeasible or gradient calculation failed
    if val == np.inf or g is None:
         if debug: print(f"Outer Opt: Initial state infeasible (H=inf) or gradient calculation failed. Stopping.")
         # The optimization effectively fails from the start
         # Return the initial w and inf H*, None details, and 0 iterations
         # Retain the initial w in best_w even if H is inf, as it's the point where failure occurred.
         best_w = w.copy()
         return np.inf, best_w, None, None, None, 0

    # Update best state with initial valid state
    best_H = val
    best_w = w.copy()
    best_pi = pi
    best_lam = lam
    best_subset = subset
    if debug: print(f"Outer Opt: Initial H(w) = {val:.6e}")


    # Store the learning rate that will be used to *start* the line search in the *next* iteration.
    # It's updated based on the success/failure of the *current* iteration's accepted step.
    current_lr_for_next_iter = lr_init # Start with the initial lr value

    # --- Projected Gradient Ascent Loop ---
    for it in range(1, MAX_OPT_ITERS + 1): # Use global MAX_OPT_ITERS here
        if debug: print(f"\nOuter Opt: --- Iteration {it} ---")
        if debug: print(f"Outer Opt: Current w = {w}")
        # Use the value and gradient from the current point 'w' obtained from the previous accepted step
        # Current state: w, val, g, pi, lam, subset

        # Convergence check: norm of the projected gradient
        # The projected gradient is proj_simplex(w + g) - w
        # If its norm is close to zero, w is a stationary point on the simplex.
        # Check using the gradient *at the current point w* (which is stored in variable 'g')
        w_plus_g = w + g
        w_projected_for_grad_check = proj_simplex(w_plus_g)
        # The actual projected gradient vector is the difference between the projected point and the current point
        # This vector points from w towards the best point on the simplex in the gradient direction.
        projected_grad_vector = w_projected_for_grad_check - w
        projected_grad_norm = np.linalg.norm(projected_grad_vector)

        if debug: print(f"Outer Opt: Projected gradient norm = {projected_grad_norm:.2e} (tolerance = {tol_grad:.2e})")

        # Use the relaxed tolerance for projected gradient norm check
        if projected_grad_norm < tol_grad:
            if debug: print(f"Outer Opt: Projected gradient norm below tolerance. Converged at iter {it}.")
            converged_iters = it # Record iteration count
            break # Stop optimization loop

        # --- Line Search ---
        # Start the line search with the learning rate value decided in the previous iteration (or initial lr).
        lr_current_step = current_lr_for_next_iter
        if debug: print(f"Outer Opt: Starting line search with lr = {lr_current_step:.2e}")

        found_accepted_step = False
        step_was_numerically_zero = False # Flag to check if step size was numerically zero

        # Store the initial value for the line search improvement check
        val_at_start_of_line_search = val

        ls_step_count = 0 # Reset line search step counter for each outer iteration

        while lr_current_step >= min_lr_line_search and ls_step_count < max_line_search_steps:
            ls_step_count += 1 # Increment line search step counter

            # Calculate the potential new point
            w_potential = w + lr_current_step * g

            # Project the potential point back onto the simplex
            w_new = proj_simplex(w_potential)

            # Check if the projected step results in a numerically significant change in w
            # This helps avoid getting stuck at boundaries or with tiny numerical steps
            step_size_norm = np.linalg.norm(w_new - w)
            if debug: print(f"  Line Search: Trying lr={lr_current_step:.2e} (step {ls_step_count}/{max_line_search_steps}). Step size norm = {step_size_norm:.2e} (threshold={step_size_norm_threshold:.2e})")

            # Use the adjusted step size norm threshold
            if step_size_norm < step_size_norm_threshold:
                 step_was_numerically_zero = True
                 if debug: print(f"  Line Search: Step size norm below threshold. No significant move on simplex. Stopping line search.")
                 break # Exit line search; no effective step found with this LR

            # Evaluate the objective function and gradient at the new point w_new
            # Pass down tolerances and debug flag
            # Avoid printing debug info from H_and_grad *within* every line search step unless really needed
            val_new, g_new, pi_new, lam_new, subset_new = H_and_grad(w_new, R, Sigma, tol_inner_feas, rcond_inner_pinv, debug=False) # debug=False here

            # If evaluation at the new point failed (inner problem infeasible or gradient bad)
            # This w_new is not valid. Reduce learning rate and try again.
            if val_new == np.inf or g_new is None:
                 if debug: print(f"  Line Search: Eval at w_new failed. val_new={val_new}. Reducing lr_current_step.")
                 lr_current_step *= lr_decay
                 continue # Backtracking: try a smaller step with reduced lr_current_step

            # Check if the new point improves the objective function (H(w) should increase for maximization)
            # Accept step if value does not decrease significantly (using tolerance)
            # Check against the value at the start of the line search (val_at_start_of_line_search)
            if val_new >= val_at_start_of_line_search - tol_val_improve:
                # Step accepted!
                if debug: print(f"  Line Search: Step accepted. val_new ({val_new:.6e}) >= val_start ({val_at_start_of_line_search:.6e}) - tolerance ({tol_val_improve:.2e}).")
                # Update the current solution point and its corresponding values
                w = w_new
                val = val_new # Update 'val' to the accepted value for the *next* iteration's line search start
                g = g_new # Gradient at the new point w_new (will be used in the *next* iteration's projected gradient check and step calculation)
                pi, lam, subset = pi_new, lam_new, subset_new # Update inner solution details

                # Update the overall best state found so far
                if val > best_H:
                     best_H = val
                     best_w = w.copy()
                     best_pi = pi
                     best_lam = lam
                     best_subset = subset
                     if debug: print(f"  Line Search: New best H found: {best_H:.6e}")


                # Update the learning rate that will be used to start the line search in the *next* iteration.
                # If a step was successful with lr_current_step, we are optimistic and try a larger starting lr next time.
                current_lr_for_next_iter = min(lr_current_step * lr_grow, max_lr_next_iter)
                found_accepted_step = True
                break # Exit line search, a step was successfully accepted

            else:
                # Step rejected, value decreased significantly
                if debug: print(f"  Line Search: Step rejected. val_new ({val_new:.6e}) < val_start ({val_at_start_of_line_search:.6e}) - tolerance ({tol_val_improve:.2e}). Reducing lr_current_step.")
                # Decrease learning rate for the *next* line search attempt within *this* iteration.
                lr_current_step *= lr_decay


        # --- End of Line Search ---

        # If line search failed to find any accepted step (lr_current_step dropped below min_lr or step_was_zero)
        if not found_accepted_step:
             if debug:
                 if step_was_numerically_zero:
                     print(f"Outer Opt: Line search resulted in numerically zero step at iter {it} (threshold={step_size_norm_threshold:.2e}). Stopping optimization.")
                 elif ls_step_count >= max_line_search_steps:
                     print(f"Outer Opt: Line search reached max steps ({max_line_search_steps}) at iter {it}. Stopping optimization.")
                 else: # lr_current_step < min_lr_line_search
                     print(f"Outer Opt: Line search failed (lr < min_lr_line_search) at iter {it}. Stopping optimization.")


             # Optimization stops. The best_H, best_w etc. variables hold the best valid state found *up to the start of this iteration*.
             converged_iters = it # Record iteration count where failure occurred (or stopped)
             break # Exit gradient ascent loop


    # --- After loop (either converged, max_iters reached, or line search failed) ---
    if debug: print(f"\nOuter Opt: --- Optimization Finished ---")
    if it == MAX_OPT_ITERS: # Check the loop variable 'it' to see if max_iters was reached
         if debug: print(f"Outer Opt: Reached max iterations {MAX_OPT_ITERS}.")
         # converged_iters is already MAX_OPT_ITERS by default

    # Return the best valid state found during optimization (best_H, best_w, etc.).
    # This handles cases where optimization stopped before convergence but might have found a better point than the initial one.

    if best_H == -np.inf: # This means no valid step was ever accepted, likely initial state was inf
         if debug: print(f"Outer Opt: Returning infeasible state after loop.")
         return np.inf, best_w, None, None, None, converged_iters # best_w might still be initial w

    # Re-evaluate H_and_grad at the best_w found to get the corresponding pi, lam, subset accurately.
    # This ensures the returned pi, lam, subset exactly match the returned H_star and w_star.
    # Use the same tolerances as during optimization, enable debug for final eval if desired.
    H_star_final, g_final, pi_star_final, lam_star_final, subset_star_final = H_and_grad(best_w, R, Sigma, tol_inner_feas, rcond_inner_pinv, debug=False) # Debug off for final eval

    # Just to be safe, ensure H_star_final is not inf (should match best_H)
    if H_star_final == np.inf:
         if debug: print(f"Outer Opt: Final H_and_grad at best_w={best_w} returned inf during final evaluation. This is unexpected.")
         return np.inf, best_w, None, None, None, converged_iters

    # Return the best found state and the iteration count where it stopped
    return H_star_final, best_w, pi_star_final, lam_star_final, subset_star_final, converged_iters

```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
run_alpha_sweep_debug.py  (FIXv2)
----------------------------------
 imports Basic_con_fun
 sweeps grid (0.02.0, 101pts)
 calls build_params   optimise_outer(debug=True)
 logs optimiser stdout/stderr
 writes alpha_sweep_results.csv + alpha_sweep_debug.log
"""

import sys
import io
import logging
from datetime import datetime
from contextlib import redirect_stdout, redirect_stderr

import numpy as np
import pandas as pd

# ----------------------------------------------------------------------
# 1.  Shared experiment library
# ----------------------------------------------------------------------
try:
    import Basic_con_fun as bc
except ModuleNotFoundError as e:
    sys.stderr.write("ERROR: Basic_con_fun not found in PYTHONPATH.\n")
    raise e


# ----------------------------------------------------------------------
# 2.  Sweep settings & logger
# ----------------------------------------------------------------------
ALPHA_MIN, ALPHA_MAX, N_ALPHA = 0.0, 2.0, 101
ALPHA_GRID = np.linspace(ALPHA_MIN, ALPHA_MAX, N_ALPHA)
LOG_FILE = "alpha_sweep_debug.log"

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s  [%(levelname)s]  %(message)s",
    handlers=[logging.FileHandler(LOG_FILE, mode="w"),
              logging.StreamHandler(sys.stdout)],
)
logger = logging.getLogger(__name__)


# ----------------------------------------------------------------------
# 3.  Helpers
# ----------------------------------------------------------------------
def pad(vec, length):
    out = np.full(length, np.nan)
    if vec is not None:
        out[: len(vec)] = vec
    return out


def support_indices(vec, tol=1e-8):
    if vec is None:
        return []
    return [int(j) for j, x in enumerate(vec) if x > tol]


# ----------------------------------------------------------------------
# 4.  Main sweep
# ----------------------------------------------------------------------
def main():
    K, M = bc.K, bc.M
    logger.info("========== Sweep START (K=%d, M=%d) ==========", K, M)
    t0 = datetime.now()

    col_w = [f"w*_m{m}" for m in range(M)]
    columns = ["alpha", "H_star", "supp_w", "lambda_min",
               "iterations", "line_search_log"] + col_w
    rows_r, rows_sig, rows_cor = [], [], []
    rows = []

    for idx, alpha in enumerate(ALPHA_GRID, start=1):
        logger.info("-----  = %.5f  (%d/%d) -----", alpha, idx, N_ALPHA)

        # 4.1 Build parameters (capture min for debug)
        R, , Corr_mats, _log = bc.build_params(alpha, return_lambda_log=True, return_corr=True)
        _min = float(np.min(_log)) if _log else np.nan
        # --- store permodel parameters ---
        for m in range(M):
          # r^m
          for k in range(K):
            rows_r.append((alpha, m, k, float(R[k, m])))
          # sigma^m
          sig_vec = bc.sigma_base + alpha * bc.Delta_sigma_m[:, m]
          for k in range(K):
            rows_sig.append((alpha, m, k, float(sig_vec[k])))
          # corr^m ( i<=j)
          C = Corr_mats[m]
          for i in range(K):
            for j in range(i, K):
              rows_cor.append((alpha, m, i, j, float(C[i, j])))

        # 4.2 Run optimiser, capture verbose output
        buf = io.StringIO()
        with redirect_stdout(buf), redirect_stderr(buf):
            H_star, w_star, pi_star, lam_star, subset_star, iters = (
                bc.optimise_outer(R, , debug=True)
            )
        line_log = buf.getvalue().replace("\n", " | ")

        # 4.3 Collect record
        row = [
            alpha,
            H_star,
            support_indices(w_star),
            _min,
            iters,
            line_log,
        ] + pad(w_star, M).tolist()

        rows.append(row)

    # ------------------------------------------------------------------
    # 5.  Write CSV
    # ------------------------------------------------------------------
    df = pd.DataFrame(rows, columns=columns)
    csv_path = "alpha_sweep_results.csv"
    df.to_csv(csv_path, index=False)
    pd.DataFrame(rows_r,   columns=["alpha", "model", "asset", "r"]
            ).to_csv("param_returns.csv", index=False)
    pd.DataFrame(rows_sig, columns=["alpha", "model", "asset", "sigma"]
            ).to_csv("param_sigmas.csv", index=False)
    pd.DataFrame(rows_cor, columns=["alpha", "model", "i", "j", "corr"]
            ).to_csv("param_corrs.csv", index=False)
    logger.info("Parameter CSVs saved: param_returns.csv, param_sigmas.csv, param_corrs.csv")
    logger.info("Results saved to %s (rows=%d)", csv_path, len(df))

    # ------------------------------------------------------------------
    # 6.  Finish
    # ------------------------------------------------------------------
    elapsed = datetime.now() - t0
    logger.info("========== Sweep finished in %s ==========", elapsed)
    print(f"\nFinished sweep: {len(df)} points  {csv_path}")


# ----------------------------------------------------------------------
# 5.  CLI entry
# ----------------------------------------------------------------------
if __name__ == "__main__":
    try:
        main()
    except Exception as exc:                # pragma: no cover
        logger.exception("Fatal error")
        sys.exit(1)
```
